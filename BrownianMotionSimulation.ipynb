This Project Simulates Brownian Motion .
To do this first i import Required libraries

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d as p3
from matplotlib.animation import FuncAnimation as animation
%matplotlib inline
```

I will use numpy for arraying my data and getting random data . for Data Visualization i will use pyplot from matplotlib.

```python
npoints = 1000
T = 1.0
```

First I determine in npoints how many points i will have to move my particle which i preferred 1000 .  Then for T we have our timeframe . it means that our particle will move in 1.0s .

```python
xaxis = np.linspace(0.0,T,npoints)
print(xaxis[1])
```

With this line, I divide my timeframe to npoints parts . which is 0.001001001001001001 for each part.

```python
dT = xaxis[1] - xaxis[0]
print(dT)
```

in this line we get timesteps between each time that our object will move .

```python
dX = np.sqrt(dT) * np.random.randn(1,npoints)
```

For brownian move we have Xt2 - Xt1 ~ Normally distributed with mean 0 and variance t2 - t1.


instead of finding an specific and single formula for our brownian motion , i generate these various normally distributed increments.

```python
X = np.cumsum(dX,axis = 1)
```

Then i get cumulative sum of dX

```python
plt.plot(xaxis , X[0,:])
plt.show()
```

so we plotted a brownian motion path graph in timeframe of 1s

```python
nparticle = 10
dX = np.sqrt(dT) * np.random.randn(nparticle,npoints)
X = np.cumsum(dX,axis = 1)
for i in range(nparticle):
    plt.plot(xaxis , X[i,:])
plt.show()
```

also we can plot several particle brownian motion.

```python
nparticle = 1000
dX = np.sqrt(dT) * np.random.randn(nparticle,npoints)
X = np.cumsum(dX,axis = 1)
for i in range(nparticle):
    plt.plot(xaxis , X[i,:])
plt.show()
```

I tried to plot brownian motion path for 1000 particle . thing which is intersting to me is even though we had totally independent increments, we get some ordered shape. all of our increments were generated randomly but we can see a parabolic shape appears.

## 2D Brownian Motion

Also we can simulate brownian motion in two dimension. to do that we need to replace our xaxis (time) on our graph with Yaxis of our movement but first we need to generate random increments for Yaxis too .

```python
dX = np.sqrt(dT) * np.random.randn(1,npoints)
X = np.cumsum(dX,axis = 1)
dY = np.sqrt(dT) * np.random.randn(1,npoints)
Y = np.cumsum(dY,axis = 1)
```

```python
plt.plot(X[0,:],Y[0,:])
plt.show()
```

as we can see we plotted path of a particle in brownian motion

```python
nparticle = 5
dX = np.sqrt(dT) * np.random.randn(nparticle,npoints)
X = np.cumsum(dX,axis = 1)
dY = np.sqrt(dT) * np.random.randn(nparticle,npoints)
Y = np.cumsum(dY,axis = 1)
for i in range(nparticle):
    plt.plot(X[i,:] , Y[i,:])
plt.show()
```

in these lines we managed to plot  2D path of several particles Brownian motion.
what is exciting for me is this looks exactly like diffusion of colorful drops of ink in water.

## 3D Brownian Motion

To simulate brownian motion in 3 dimension we need an extra z dimension with random increments


```python
dX = np.sqrt(dT) * np.random.randn(1,npoints)
X = np.cumsum(dX,axis = 1)
dY = np.sqrt(dT) * np.random.randn(1,npoints)
Y = np.cumsum(dY,axis = 1)
dZ = np.sqrt(dT) * np.random.randn(1,npoints)
Z = np.cumsum(dZ,axis = 1)
sp = plt.axes(projection = "3d")
sp.plot3D(X[0,:] , Y[0,:] , Z[0,:] )
plt.show()
```

same as before we can plot Brownian Motion graph for several particles.

```python
nparticle = 3
dX = np.sqrt(dT) * np.random.randn(nparticle,npoints)
X = np.cumsum(dX,axis = 1)
dY = np.sqrt(dT) * np.random.randn(nparticle,npoints)
Y = np.cumsum(dY,axis = 1)
dZ = np.sqrt(dT) * np.random.randn(nparticle,npoints)
Z = np.cumsum(dZ,axis = 1)
sp = plt.axes(projection = "3d")
for i in range(nparticle):
    sp.plot3D(X[i,:] , Y[i,:] , Z[i,:] )
plt.show()
```

Now something intersting
